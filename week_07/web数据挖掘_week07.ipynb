## 3. 构建参数模板

```python
def parse_url_qs(url):
    six_parts = urlparse(url)
    out = parse_qs(six_parts.query)
    return out

参数模板 = parse_url_qs(list(公司数据选择器.values())[0])
print(参数模板)

dict_comp = {k:parse_url_qs(v)["compTag"][0] for k,v in 公司数据选择器.items()}
print(dict_comp)
```

```python
dict_industry = {k:parse_url_qs(v)["industries"][0] for k,v in 行业数据选择器.items()}
print(dict_industry)
```

```python
dict_city = {k:parse_url_qs(v)["dqs"][0] for k,v in 城市数据选择器.items()}
print(dict_city)
```

```python
# 公司
def 参数模板生成(comp,keyword):
    参数 = 参数模板.copy()
    参数['compTag'] = comp
    参数['keyword'] = keyword
    return(参数)

参数_comp_python = {k:参数模板生成(comp = [v],keyword = ["python"]) for k,v in dict_comp.items()}
print(参数_comp_python)
```

```python
# 行业
def 参数模板生成(industries,keyword):
    参数 = 参数模板.copy()
    参数['industries'] = industries
    参数['keyword'] = keyword
    return(参数)

参数_industries_python = {k:参数模板生成(industries = [v],keyword = ["python"]) for k,v in dict_industry.items()}
print(参数_industries_python)
```

```python
# 城市
def 参数模板生成(city,keyword):
    参数 = 参数模板.copy()
    参数['dqs'] = city
    参数['keyword'] = keyword
    return(参数)

参数_city_python = {k:参数模板生成(city = [v],keyword = ["python"]) for k,v in dict_city.items()}
print(参数_city_python)
```

## 4. requests生成

```python
session = HTMLSession()

def requests_liepin( url,params):
    r = session.get(url,params = payload)
    _main = r.html.xpath("//ul[@class='sojob-list']/li")
    dict_xpath = {
        "text":{
            "学历": '//div[contains(@class,"job-info")]/p/span[@class="edu"]',
            "经验": '//div[contains(@class,"job-info")]/p/span[@class="edu"]/following-sibling::span',
            "薪水": '//div[contains(@class,"job-info")]/p/span[@class="text-warning"]', 
            "时间": '//div[contains(@class,"job-info")]/p/time/@title', 
            "职称": '//div[contains(@class,"job-info")]/h3/a', 
            "公司地点": '//div[contains(@class,"job-info")]/p/a',
            "公司名称": '//div[contains(@class,"sojob-item-main")]//p[@class="company-name"]/a',            
        },
        "text_content":{
        },
        "href":{
            "链接": '//div[contains(@class,"job-info")]/h3/a',
            "公司url": '//div[contains(@class,"sojob-item-main")]//p[@class="company-name"]/a',
        }
    }
    
    def get_text_content(_xpath_):
        result = [e.xpath(_xpath_)[0].lxml.text_content() for e in _main]
        return(result)
    def get_text(_xpath_):
        result = ["".join([x.strip() if type(x) is str else x.text.strip() for x in e.xpath(_xpath_)]) for e in _main]
        return(result)
    def get_e_href(_xpath_):
        result = [list(e.xpath(_xpath_, first=True).absolute_links)[0] \
                   if len(e.xpath(_xpath_, first=True).absolute_links) >= 1  \
                   else "" for e in _main]
        return(result)
    
    数据字典 = dict()

    数据字典 = {k:get_text_content(v) for k,v in dict_xpath['text_content'].items()}
    数据字典.update({k:get_text(v) for k,v in dict_xpath['text'].items()})
    数据字典.update({k:get_e_href(v) for k,v in dict_xpath['href'].items()})

    数据 = pd.DataFrame(数据字典)
    #数据.to_excel("20春_Web数据挖掘_week03_liepin.xlsx", sheet_name="搜查结果")
    return (数据)
```

## 5. 翻页
### (1) xpath解析翻页a/@href

```python
url = "https://www.liepin.com/zhaopin/?keyword=Python"
session = HTMLSession()
payload = 参数_industries_python['互联网/电商']
r = session.get( url,params = payload )
```

```python
xpath_翻页_comp = "//div[@class='pagerbar']/a[starts-with(@href,'/zhaopin')]"
print(r.html.xpath(xpath_翻页_comp))

href_list = [x.xpath('//@href')[0] for x in r.html.xpath(xpath_翻页_comp)]
# print(href_list)

word_list = [x.text for x in r.html.xpath(xpath_翻页_comp)]
# print(word_list)

word_href_dict = {x.text:x.xpath('//@href')[0] for x in r.html.xpath(xpath_翻页_comp)}
print(word_href_dict)
```

### (2) 构建参数模板

```python
from urllib.parse import urlparse,parse_qs
from IPython.display import display,HTML

df = pd.DataFrame([urlparse(x) for x in href_list])
# display(df)
# print(df.nunique())

df_qs = pd.DataFrame([{k:v[0] for k,v in parse_qs(x).items()}for x in df["query"]])
display(df_qs)
print(df_qs.nunique())

df_qs = df_qs.assign(curPage_int = df_qs.curPage.astype(int))
df_qs[["curPage","compTag","industries","keyword"]]
```

```python
def parse_url_curPage(url):
    six_parts = urlparse(url)
    out =parse_qs(six_parts.query)
    return(out)

参数模板_curPage = parse_url_curPage(href_list[0])
print(参数模板_curPage)
print(word_href_dict)
```

```python
# 构建模板生成器
def 参数模板生成_curpage(keyword,curPage):
    参数 = 参数模板_curPage.copy()
    参数["curPage"] = curPage
    参数["keyword"] = keyword
    return(参数)

参数_python_curPage = {k:参数模板生成_curpage(curPage=[v],keyword=["python"]) for k,v in word_href_dict.items()}
print(参数_python_curPage)

print(df_qs.curPage_int.min())
print(df_qs.curPage_int.max())

参数_python_curPage = {i:参数模板生成_curpage(curPage=[i],keyword=["python"]) for i in range(0,df_qs.curPage_int.max()+1)}
参数_python_curPage
```

### (3) 开始翻页

```python
# 一个关键词
import time
from random import random
url = "https://www.liepin.com/zhaopin/?keyword=Python"
list_python = list()
for k,v in 参数_python_curPage.items():
    payload = v
    df = requests_liepin( url, params = payload)
    time.sleep(3+4*random())
    df = df.assign (curPage = k)
    df = df.assign (compTag = "中国500强")
    df = df.assign (industries = "互联网/电商")
    list_python.append(df)
    
df_all = pd.concat(list_python).reset_index()
df_all.index.name = '序'
# display(df_all)
df_all.to_excel("python_翻页.xlsx",sheet_name="python")
```
